```python
#  create list comprehension to include only movies with a Director and imdb link
wiki_movies = [movie for movie in wiki_movies_raw
              if ('Director' in movie or 'Directed by' in movie)
              and 'imdb_link' in movie
              and 'No. of episodes' not in movie]
len(wiki_movies)
```

```python
# Create new df from wiki_movies list comprehension
wiki_movies_filtered_df = pd.DataFrame(wiki_movies)
wiki_movies_filtered_df.head()
```

```python
# create function to clean movie data
def clean_movie(movie):
    movie = dict(movie) #Create a non-destructive copy
    alt_titles = {} # create empty dictionary to hold alternative titles
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:   # Loop through the list of all alternative titles
        if key in movie:                                 # Check to see if current key exists in movie
            alt_titles[key] = movie[key]                 # If there, remove key-value pair and add to alternative titles dictionary  
            movie.pop(key)
    if len(alt_titles) > 0:                              # after loop, add alternative titles dict to the movie object
        movie['alt_titles'] = alt_titles
            
    return movie
```

```python
# make a list of cleaned movies with list comprehension
clean_movies = [clean_movie(movie) for movie in wiki_movies]
```
